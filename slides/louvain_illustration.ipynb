```python
import networkx as nx
import matplotlib.pyplot as plt

# Build karate club graph
G_karate, _ = build_graph("karate", seed=42)
pos_karate = compute_pos_orig(G_karate, seed=42)

# Create the plot
plt.figure(figsize=(10, 8))
nx.draw(
    G_karate, 
    pos_karate, 
    node_color='black', 
    edge_color='grey', 
    node_size=300,
    with_labels=True,
    font_color='white',
    font_size=8
)
#plt.title("Karate Club Graph")
plt.savefig('images/karate_club_graph.png', dpi=300, transparent=False, bbox_inches='tight')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Print edge list and node properties for karate club graph
print("## Karate Club Graph - Edge List (first 10 edges)")
print("| Src | Tgt | Wgt |")
print("|-----|-----|-----|")
edge_list = list(G_karate.edges(data=True))
for i, (u, v, data) in enumerate(edge_list[:10]):
    weight = data.get('weight', 1.0)
    print(f"| {u:^3} | {v:^3} | {weight:^3.1f} |")

print(f"\n**Total edges:** {len(edge_list)}")

print("\n## Node Properties")
print("| Node | Deg | Clust |")
print("|------|-----|-------|")
clustering = nx.clustering(G_karate)
for node in sorted(G_karate.nodes())[:10]:  # Show first 10 nodes
    degree = G_karate.degree(node)
    cc = clustering[node]
    print(f"| {node:^4} | {degree:^3} | {cc:^5.3f} |")

print(f"\n**Total nodes:** {G_karate.number_of_nodes()}")
print(f"**Graph density:** {nx.density(G_karate):.3f}")
print(f"**Average clustering coefficient:** {nx.average_clustering(G_karate):.3f}")
```

```python
# Print adjacency matrix for karate club graph
import pandas as pd

print("## Karate Club Graph - Adjacency Matrix")
print()

# Get adjacency matrix as numpy array
adj_matrix = nx.adjacency_matrix(G_karate).todense()

# Convert to pandas DataFrame for nice display
adj_df = pd.DataFrame(adj_matrix, 
                     index=sorted(G_karate.nodes()), 
                     columns=sorted(G_karate.nodes()))

# Create a view showing first 5 and last 5 rows/columns with ellipsis in middle
n_nodes = adj_df.shape[0]
if n_nodes > 10:
    # Get first 5 and last 5 rows/columns
    top_left = adj_df.iloc[:5, :5]
    top_right = adj_df.iloc[:5, -5:]
    bottom_left = adj_df.iloc[-5:, :5]
    bottom_right = adj_df.iloc[-5:, -5:]
    
    # Create ellipsis columns for the middle
    ellipsis_col_top = pd.DataFrame(['...'] * 5, index=top_left.index, columns=['...'])
    ellipsis_col_bottom = pd.DataFrame(['...'] * 5, index=bottom_left.index, columns=['...'])
    
    # Create ellipsis row for the middle
    all_cols = list(top_left.columns) + ['...'] + list(top_right.columns)
    ellipsis_row = pd.DataFrame([['...'] * len(all_cols)], index=['...'], columns=all_cols)
    
    # Combine top half
    top_half = pd.concat([top_left, ellipsis_col_top, top_right], axis=1)
    
    # Combine bottom half  
    bottom_half = pd.concat([bottom_left, ellipsis_col_bottom, bottom_right], axis=1)
    
    # Combine all parts
    display_df = pd.concat([top_half, ellipsis_row, bottom_half])
    
    # Display and print markdown table
    display(display_df)
    
    # Print as markdown table
    print("\n### Adjacency Matrix (Markdown format)")
    print(display_df.to_markdown())
else:
    # Display and print markdown table
    display(adj_df)
    
    # Print as markdown table
    print("\n### Adjacency Matrix (Markdown format)")
    print(adj_df.to_markdown())

print(f"\n(Full matrix is {adj_df.shape[0]}x{adj_df.shape[1]})")
```
