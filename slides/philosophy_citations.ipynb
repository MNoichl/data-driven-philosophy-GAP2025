```python
import matplotlib.pyplot as plt


plt.style.use("opinionated_rc")

pd.set_option('display.max_columns', None)

graphi_dir = 'images'
```

```python
philosophy_base_df = pd.read_pickle('/Users/Noich001/Desktop/philorisk_hein/data_sets/philosophy_base_df.pkl')
```

```python
plt.figure(figsize=(10, 6))
data = philosophy_base_df[philosophy_base_df['publication_year'] == 1990][philosophy_base_df['cited_by_count'] < 100]['cited_by_count']
ax = sns.histplot(data, bins=range(0, 101), alpha=0.8, color='darkred')
ax.set_xlabel('Citations', fontsize=12)
ax.set_ylabel('Count (truncated at 100+ citations)', fontsize=12)
ax.set_title('Citation Distribution for Philosophy Papers (1990)')
# ax.grid(True, alpha=0.3)


plt.tight_layout()
plt.savefig('images/citations_distribution.png', dpi=300, transparent=False, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
citation_counts = philosophy_base_df['cited_by_count'].value_counts().sort_index()
plt.loglog(citation_counts.index, 
           citation_counts.values, 
           'o-', alpha=0.7, markersize=3)
plt.xlabel('Citations')
plt.ylabel('Count')
plt.title('Citation Distribution (Log-Log Plot)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
